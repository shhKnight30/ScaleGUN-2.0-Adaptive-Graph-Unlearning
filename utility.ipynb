```python

import re
import pandas as pd
import  numpy as np
# Initialize lists to store extracted values
iterations = []
test_acc = []
avg_update_cost = []
avg_unlearn_cost = []
avg_tot_cost = []

# Regular expression pattern to match the lines
pattern = re.compile(
    r"Iteration (\d+):.*Test acc = ([\d.]+), avg update cost: ([\d.]+), avg unlearn cost:([\d.]+), avg tot cost:([\d.]+)"
)

# Read the log file
with open(scalegun_file, 'r') as file:
    for line in file:
        match = pattern.search(line)
        if match:
            iteration = int(match.group(1))
            if 0 <= iteration <= 100:
                # Extract values and append to lists
                iterations.append(iteration)
                test_acc.append(float(match.group(2)))
                avg_update_cost.append(float(match.group(3)))
                avg_unlearn_cost.append(float(match.group(4)))
                avg_tot_cost.append(float(match.group(5)))

# Create a DataFrame from the lists
data = pd.DataFrame({
    'Iteration': iterations,
    'Test Accuracy': test_acc,
    'Average Update Cost': avg_update_cost,
    'Average Unlearn Cost': avg_unlearn_cost,
    'Average Total Cost': avg_tot_cost
})

# Save the DataFrame to a CSV file
data.to_csv('extracted_data.csv', index=False)
np.savetxt("ScaleGUN_products.txt",test_acc)
```
